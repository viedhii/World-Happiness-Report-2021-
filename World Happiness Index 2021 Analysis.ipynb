```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\bhara\AppData\Local\Programs\Python\Python38\Scripts\world_happiness_report_2021.csv')
df.head()
```

```python
df.tail(5)
```

```python
sns.distplot(df["Ladder score"])
```

```python
df.hist(figsize=(10, 8))
```

```python
plt.figure(figsize=(30,10))
plt.scatter(df['Country name'],df['Ladder score'],c='blue') 
plt.xlabel('Country name',fontsize=18) 
plt.ylabel('Ladder score',fontsize=18)
plt.xticks(rotation=30)
plt.title('Scatterplot of Country name vs Ladder score ') 
plt.show() 
```

```python

plt.subplots(figsize = (25,10))
plt.xticks(rotation=30)
sns.boxplot(data = df,x ='Country name', y = 'Logged GDP per capita')
```

```python
plt.subplots(figsize = (25,10))
plt.xticks(rotation=30)
sns.barplot(data = df,x = 'Regional indicator', y = 'Healthy life expectancy') 
```

```python
sns.pairplot(df[['Social support','Healthy life expectancy']])
```

```python
corr = df.corr()
plt.figure(figsize=(14,7))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='Explained by: Social support',y='Explained by: Healthy life expectancy',data=df)
plt.ylim(0,)
```

```python
plt.figure(figsize=(6,5))
sns.regplot(x='Ladder score',y='Logged GDP per capita',data=df)
plt.ylim(0,)
```

```python
After visualizing the dataset we can infere that countries with high ladder score had better logged gdp income and 
healthy life expentency as compared to the once with low ladder score.
```
